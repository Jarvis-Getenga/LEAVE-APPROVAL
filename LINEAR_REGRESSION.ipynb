```python
#DEVELOPING MODEL FOR ESTIMATING HEATING OIL USED FOR A FAMILY IN JANUARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#DATASET
data = {
    'Oil': [275.3,363.8,164.3,40.8,94.3,230.9,366.7,300.6,
            237.8,121.4,31.4,203.5,441.1,323.0,52.5],

    'Temp': [40,27,40,73,64,34,9,8,23,63,65,41,21,38,58],

    'Insulation': [3,3,10,6,6,6,6,10,10,3,10,6,3,3,10]
}

df = pd.DataFrame(data)

df

#TARGET
X = df[['Temp', 'Insulation']]
y = df['Oil']

model = LinearRegression()

model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

#PREDICTIONS
prediction = model.predict([[15,10]])

print("Predicted Oil Consumption:", prediction[0])

#VISUALIZATION
plt.figure(figsize=(7,5))
plt.scatter(df['Temp'], df['Oil'])
plt.xlabel("Temperature")
plt.ylabel("Oil Consumption")
plt.title("Temperature vs Oil Consumption")
plt.grid(True)
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(df['Insulation'], df['Oil'])
plt.xlabel("Insulation")
plt.ylabel("Oil Consumption")
plt.title("Insulation vs Oil Consumption")
plt.grid(True)
plt.show()
```

A model was developed to estimate the amount of heating oil used by a single-family home during January using:

Average outside temperature
Amount of attic insulation

The model used was a Multiple Linear Regression Model.

The regression equation is:

Y^=β0​+β1​T+β2​I

Where:

Y^
 = Estimated heating oil consumption (gallons)

T = Average outside temperature (°F)

I = Insulation thickness (inches)

1. Heating oil consumption increases when temperature decreases

The data shows that homes consume more heating oil during colder temperatures.

For example:

At very low temperatures such as 8°F and 9°F, oil consumption exceeded 300 gallons.
At higher temperatures such as 64°F and 73°F, oil consumption dropped below 100 gallons.

This indicates an inverse relationship between temperature and oil consumption.

2. Better insulation reduces heating oil usage

Homes with thicker insulation generally used less heating oil.

This is because insulation helps retain heat inside the house, reducing the amount of heating required.

3. Temperature has the greatest influence on oil consumption

Although insulation affects heating efficiency, the outside temperature appears to have the strongest impact on the amount of oil used.

Large changes in oil consumption were observed whenever temperature changed significantly.

4. Multiple Linear Regression is appropriate for the model

The relationships between:

Temperature and oil consumption
Insulation and oil consumption

appear reasonably linear, making multiple linear regression suitable for prediction.
