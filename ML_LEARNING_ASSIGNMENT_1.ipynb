```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file = pd.read_csv('/iebc_2017_presidential_results.csv')

# Data for pie chart
parties = file['Political Party / Independent']
percentages = file['Percentage of Votes Cast'].str.replace('%', '').astype(float)

# Create pie chart
plt.figure(figsize=(10,10))
plt.pie(percentages, labels=parties, autopct='%1.1f%%', startangle=90)
plt.title('2017 Presidential Election Vote Percentage by Political Party')
plt.show()
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

# Generate Fibonacci numbers
fib = [0, 1]
for i in range(8):
    fib.append(fib[-1] + fib[-2])

# X-axis values
x = range(1, 11)

# Continuous line plot
plt.figure(figsize=(8,5))
plt.plot(x, fib, marker='o', linestyle='-')
plt.title('First 10 Fibonacci Numbers - Line Plot')
plt.xlabel('Position')
plt.ylabel('Fibonacci Number')
plt.grid(True)
plt.show()

# Independent dots plot
plt.figure(figsize=(8,5))
plt.scatter(x, fib)
plt.title('First 10 Fibonacci Numbers - Scatter Plot')
plt.xlabel('Position')
plt.ylabel('Fibonacci Number')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
sales = pd.read_csv('/company_sales_data.csv')

# Calculate yearly sales for each product
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

total_sales = []

for product in products:
    total_sales.append(sales[product].sum())

# Plot pie chart
plt.figure(figsize=(10,10))
plt.pie(total_sales, labels=products, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Total Units Sold Per Product')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
sales = pd.read_csv('/company_sales_data.csv')
# Extract data
months = sales['month_number']
profits = sales['total_profit']

# Styled line plot
plt.figure(figsize=(8,5))
plt.plot(months,
         profits,
         color='red',
         marker='o',
         linestyle='--',
         linewidth=3,
         markersize=8,
         label='Profit Data')

plt.title('Company Profit Per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load dataset
sales = pd.read_csv('/company_sales_data.csv')

# Data
months = sales['month_number']
facecream = sales['facecream']
facewash = sales['facewash']

# Bar positions
x = np.arange(len(months))
width = 0.35

# Create bar chart
plt.figure(figsize=(10,5))
plt.bar(x - width/2, facecream, width, label='Face Cream')
plt.bar(x + width/2, facewash, width, label='Face Wash')

# Labels and title
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Face Cream and Face Wash Sales Data')
plt.xticks(x, months)
plt.legend()
plt.grid(True)
plt.show()
```
