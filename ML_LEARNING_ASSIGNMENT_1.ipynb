```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("iebc_2017_presidential_results.csv")

# Clean percentage (remove % and convert to float)
df["Percentage"] = df["Percentage of Votes Cast"].str.replace('%','').astype(float)

plt.pie(df["Percentage"], labels=df["Candidate Name"], autopct='%1.1f%%')
plt.title("2017 Presidential Election Results")
plt.show()
```

```python
import matplotlib.pyplot as plt

fib = [0, 1]
for i in range(8):
    fib.append(fib[-1] + fib[-2])

plt.plot(fib)
plt.scatter(range(len(fib)), fib)

plt.title("Fibonacci Sequence")
plt.show()
```
